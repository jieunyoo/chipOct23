```python
def subtract1(x):
    return 1-x

df4 = df3.apply(subtract1)
df4

dfT = df4.T
```

```python
dfT
```

```python
df_xdata = pd.read_csv('voltages.csv')
xdata = df_xdata.to_numpy().flatten()
len(xdata)
xdata2 = xdata+100
xdata3 = xdata2/10
xdata3
len(xdata3)
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from scipy.stats import norm

ydata0 = dfT.iloc[0:1] #bit 0
ydata = ydata0.to_numpy().flatten()

fitResult=curve_fit(f=norm.cdf, xdata=xdata3, ydata=ydata, p0=[400,1],bounds=((-np.inf,0),(np.inf,np.inf)))
mean_, std_=fitResult[0]
print(f"Mean {mean_:.4f}, StdDev {std_:.4f}")
print('Covariance matrix:')
print(fitResult[1])
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from scipy.stats import norm


def applyFit(x):
    x.to_numpy().flatten()
    fitResult=curve_fit(f=norm.cdf, xdata=xdata3, ydata=x, p0=[400,1],bounds=((-np.inf,0),(np.inf,np.inf)))
    mean_, std_=fitResult[0]
   # print(f"Mean {mean_:.4f}, StdDev {std_:.4f}")
    x['mean'] = mean_
    x['SD'] = std_
    return x['mean'], x['SD']
```

```python
dfT[['mean','sd']]=dfT.apply(applyFit,axis=1,result_type="expand")
dfT
```

```python
dfT.to_csv('fittedData.csv',index=False)
```
